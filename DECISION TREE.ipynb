6.  Perform Decision Tree classification algorithm on given data set.

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("DecisionTreeDataSet - Sheet1.csv")
df
```

```python
# Convering datatype object into binary of clumn 'Nationality' and 'GO' such as, 'UK':0, 'USA':1 and 'N':1
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
# Feature Selection
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

print(X)
print(y)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
