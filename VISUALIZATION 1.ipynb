Use any Dataset and Visualize data using different type

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.plot(df["index"],df["V Kohli"])
```

```python
# Filter data for the years 2008 to 2012
filtered_df = df[df['index'].isin([2008, 2009, 2010, 2011, 2012])]

# Extract V Kohli's runs and years
kohli_runs = filtered_df['V Kohli']
years = filtered_df['index'].astype(str)  # Convert years to string for labels

# Creating the pie chart
plt.pie(kohli_runs, labels=years, autopct='%1.1f%%', startangle=90)
plt.title("V. Kohli's Runs (2008 - 2012)")
plt.show()
```

```python
# Filter data for the years 2008 to 2012
filtered_df = df[df['index'].isin([2008, 2009, 2010, 2011, 2012])]

# Extract data
years = filtered_df['index'].astype(str)  # Convert years to string for plotting
rg_sharma_runs = filtered_df['RG Sharma']
v_kohli_runs = filtered_df['V Kohli']

# Set the position of the bars on the x-axis
x = np.arange(len(years))

# Width of a bar
width = 0.3

# Plotting the bar graph
plt.bar(x - width/2, rg_sharma_runs, width, label="RG Sharma", color="skyblue")
plt.bar(x + width/2, v_kohli_runs, width, label="V Kohli", color="salmon")

# Adding labels and title
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs Comparison (2008 - 2012)")
plt.xticks(x, years)
plt.legend()

# Display the plot
plt.show()
```
