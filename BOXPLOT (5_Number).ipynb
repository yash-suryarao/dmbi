5 NUMBER SUMMARY USING PYTHON (Boxplot)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("Gold Price Prediction.csv")
```

```python
Price_Today_array = df["Price Today"].values
Price_Today_array
```

```python

# Sort the array
Price_Today_array = np.sort(Price_Today_array)

# Calculate Statistics
minimum = np.min(Price_Today_array) # Smallest value
q1 = np.percentile(Price_Today_array, 25) # The value below 25%
median = np.median(Price_Today_array) # Middle value
q3 = np.percentile(Price_Today_array, 75) # The value upper half of median and falls below 75%
maximum = np.max(Price_Today_array)

print(f"Minimum ={minimum}")
print(f"Q1 ={q1}")
print(f"Median ={median} ")
print(f"Q3 ={q3}")
print(f"Maximum ={maximum}")

# Plotting the boxplot
plt.boxplot(Price_Today_array,vert=False)
plt.title('Box plot of Price Today value')
plt.ylabel('Price Today')
plt.show()
```
