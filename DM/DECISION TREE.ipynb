6.  Perform Decision Tree classification algorithm on given data set.

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("Titanic.csv")
df
```

```python
# Convering datatype object into binary of clumn 'Nationality' and 'GO' such as, 'UK':0, 'USA':1 and 'N':1
d = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].map(d)
d = {'Q': 0, 'S': 1, 'C': 2}
df['Embarked'] = df['Embarked'].map(d)

print(df)
```

```python
# Feature Selection
features = ['Age', 'Survived', 'SibSp']

X = df[features]
y = df['Embarked']

print(X)
print(y)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
