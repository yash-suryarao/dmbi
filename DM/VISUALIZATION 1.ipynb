```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = 'healthcare_dataset.csv'  # Replace with your file path
hospital_data = pd.read_csv(file_path)

# Ensure column names are consistent
hospital_data.columns = [col.strip() for col in hospital_data.columns]

# Convert 'Date of Admission' to datetime
hospital_data['Date of Admission'] = pd.to_datetime(hospital_data['Date of Admission'])

# 1. Line Chart: Billing Amount over Time (by Date of Admission)
billing_by_date = hospital_data.groupby('Date of Admission')['Billing Amount'].sum()

plt.figure(figsize=(10, 6))
plt.plot(billing_by_date.index, billing_by_date.values, marker='o', color='b', linestyle='-', linewidth=2)
plt.title('Billing Amount over Time (Line Chart)', fontsize=16)
plt.xlabel('Date of Admission', fontsize=12)
plt.ylabel('Total Billing Amount', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Pie Chart: Gender Distribution
gender_counts = hospital_data['Gender'].value_counts()

plt.figure(figsize=(7, 7))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(gender_counts)))
plt.title('Gender Distribution (Pie Chart)', fontsize=16)
plt.ylabel('')
plt.show()

# 3. Bar Chart: Medical Conditions Count
condition_counts = hospital_data['Medical Condition'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=condition_counts.index, y=condition_counts.values, palette='coolwarm')
plt.title('Medical Conditions Distribution (Bar Chart)', fontsize=16)
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# 4. Pie Chart: Admission Type Distribution
admission_type_counts = hospital_data['Admission Type'].value_counts()

plt.figure(figsize=(7, 7))
admission_type_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2', len(admission_type_counts)))
plt.title('Admission Type Distribution (Pie Chart)', fontsize=16)
plt.ylabel('')
plt.show()

# 5. Bar Chart: Average Billing Amount by Admission Type
avg_billing_by_type = hospital_data.groupby('Admission Type')['Billing Amount'].mean().sort_values()

plt.figure(figsize=(8, 5))
sns.barplot(x=avg_billing_by_type.index, y=avg_billing_by_type.values, palette='plasma')
plt.title('Average Billing Amount by Admission Type (Bar Chart)', fontsize=16)
plt.xlabel('Admission Type', fontsize=12)
plt.ylabel('Average Billing Amount', fontsize=12)
plt.show()
```
