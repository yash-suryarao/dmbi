K-MEAN

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("CrimesOnWomenData.csv")
df.head()
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
X= df[['Rape', 'K&A']].values
print('Crimes Data:')
print(X)
```

```python
plt.scatter(X[:,0], X[:,1],c='black')
plt.title('Rape vs K&A')
plt.xlabel('Rape')
plt.ylabel('K&A')
plt.show()
```

```python
# perform Kmeans clustering 
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
#get the centroids (centers points of clusters ) and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
scatter =  plt.scatter (X[:, 0], X[:, 1], c= labels, cmap='rainbow')

plt.scatter(centroids [:, 0], centroids[:, 1], c='black', marker= 'X',s=100)

plt.title("Clusters Based on Rape and K&A")
plt.xlabel("Rape")
plt.ylabel("K&A")

legend_labels = [f'Cluster {1+1}' for i in range(len(centroids))] 
legend = plt.legend(handles=scatter.legend_elements() [0], labels = legend_labels, title="Clusters")

plt.show()
```
